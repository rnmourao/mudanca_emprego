```python
import h2o
from h2o.automl import H2OAutoML

try:
    h2o.cluster().shutdown()
except AttributeError:
    pass
    
h2o.init(nthreads = 6,
#          max_mem_size = "24G")         
         max_mem_size = "4G")
```

```python
my_seed = 1980
```

```python
treino = h2o.import_file('../data/treino.csv')
teste = h2o.import_file('../data/teste.csv')
```

```python
treino['label'] = treino['label'].asfactor()
teste['label'] = teste['label'].asfactor()
```

```python
# Run AutoML 
aml = H2OAutoML(max_runtime_secs=2*60*60, max_models=120, nfolds=10, stopping_metric='AUC', 
                exclude_algos=['StackedEnsemble'], seed=my_seed)
aml.train(y='label', training_frame=treino, leaderboard_frame=teste)
```

```python
lb = aml.leaderboard
```

```python
uteis = ['tempo_emprego', 'solteiro', 'tempo_empresa', 'idioma-en-fala', 'idioma-en-le', 'tempo_cargo', 
         'experiencia_meses', 'homem', 'tem_cnh', 'taxa_permanencia', 'cargo_estagiario', 'tempo_municipio', 'label']

inuteis = [x for x in treino.columns if x not in uteis]
```

```python
for i in inuteis:
    treino = treino.drop(i)
    teste = teste.drop(i)
```

```python
treino.columns
```

```python
# Run AutoML 
aml = H2OAutoML(max_runtime_secs=1*60*60, max_models=20, nfolds=10, stopping_metric='AUC', 
                exclude_algos=['StackedEnsemble'], seed=my_seed)
aml.train(y='label', training_frame=treino, leaderboard_frame=teste)
```

```python
h2o.save_model(aml.leader, path="../mdl/")
```
