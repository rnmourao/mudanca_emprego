Este notebook efetua algumas etapas mair refinadas da preparação de dados, como a identificação e tratamento de _outliers_, ou valores extremos.

Outro refinamento consiste na seleção de variáveis, uma etapa importante da construção de um modelo preditivo. Em alguns casos, variáveis muito similares podem "atrapalhar" a identificação de padrões. 

Com isso, torna-se necessário retirar variáveis que são muito parecidas entre si, de forma que os algoritmos de mineração consigam gerar modelos preditivos de melhor performance.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
# recupera arquivo de entrada
df = pd.read_csv('../data/df.csv')
```

```python
# remove campos desnecessarios
df.drop(labels=['cargo', 'empresa', 'municipio_empresa'], axis=1, inplace=True)
```

```python
# mostra estatisticas sobre cada coluna
df.describe().transpose().sort_values(by='max', ascending=False)
```

```python
# identifica pessoas com idade incompativel com o mercado de trabalho
df.loc[(df['idade'] <= 10 * 12) | (df['idade'] >= 70 * 12), ['id', 'idade']].sort_values(by='idade')
```

```python
# percentual de pessoas com idades negativas
len(np.unique(df.loc[(df['idade'] < 10 * 12), ['id']])) / float(len(np.unique(df.id)))
```

```python
# campo idade possui muitas informacoes falsas, tornando-a inutil
df.drop('idade', axis=1, inplace=True)
```

```python
# identifica variaveis numericas e nao numericas
numericos = df.select_dtypes([np.number]).columns
nao_numericos = [x for x in df.columns if x not in numericos]
nao_numericos
```

```python
# mostra distribuicao de cada variavel em um histograma, onde o eixo x corresponde ao valor da variavel e o eixo y
# eh a quantidade de registros com aquele valor
for coluna in df[numericos].columns:
    plt.figure(figsize=(7, 5))
    
    plt.hist(df[coluna])
    plt.title(coluna)
    
    plt.show()
```

```python
# cria matriz de correlacao
matriz = df[numericos].corr()
```

```python
# mostra matriz de correlacao em forma grafica
fig, ax = plt.subplots(figsize=(10,10))
ax.matshow(matriz)
plt.show()
```

```python
# mostra quais variaveis possuem maior correlacao absoluta com a variavel alvo
corr_pd = pd.DataFrame(matriz, columns=numericos)
corr_pd.index = numericos
np.abs(corr_pd['label']).sort_values(ascending=False)[:20]
```

```python
# seleciona uma amostra representativa para identificacao de multicolinearidade
amostra = df.loc[df['label'] == 1].sample(n=1000, replace=True, random_state=1980)
```

```python
# Identificar multicolinearidade usando Variance Inflation Factor (VIF)

MAX_VIF = 5
desnecessarias = nao_numericos + ['label']

achou = True
i = 0
while achou:
    i += 1
    print(i)
    
    temp_df = amostra.drop(desnecessarias, axis=1)
    temp_cols = temp_df.columns
    
    vifs = []
    for c in temp_cols:
        explicativas = [x for x in temp_cols if x != c]
        
        lr = LinearRegression(normalize=True, copy_X=True, n_jobs=3)
        modelo = lr.fit(temp_df[explicativas], temp_df[c])
        r2 = modelo.score(temp_df[explicativas], temp_df[c])
        if np.abs(r2) == 1:
            vif = np.float('inf')
        else:
            vif =  1 / float(1 - r2)
        
        vifs.append({'col' : c, 'vif' : vif})
        
        if vif == np.float('inf'):
            break
        
    vifs_pd = pd.DataFrame(vifs)
        
    idx = vifs_pd.loc[vifs_pd.vif == np.max(vifs_pd.vif)].index[0]
    k = vifs_pd.loc[idx, 'col']
    v = vifs_pd.loc[idx, 'vif']
    
    print(k + ' ' + str(v))
    
    achou = False    
    if v > MAX_VIF:
        desnecessarias.append(k)
        achou = True
```

```python
# numero de variaveis desnecessarias, de acordo com o VIF
len(desnecessarias)
```

```python
desnecessarias.sort()
desnecessarias
```

```python
# variaveis nao colineares, de acordo com VIF
vifs_pd.sort_values('vif', ascending=False)
```

```python
removiveis = [x for x in desnecessarias if x not in ['label', 'tempo_cargo', 'tempo_emprego']]
removiveis
```

```python
# remove variaveis descartadas pelo VIF
df_vif = df.drop(removiveis, axis=1)
```

```python
# colocar label como primeira coluna
colunas = df_vif.columns.tolist()
colunas.insert(0, colunas.pop(colunas.index('label')))
```

```python
# ordenacao das colunas
df_vif = df_vif[colunas]
```

```python
# nova matriz de correlacao, agora com menos variaveis
numericos = df_vif.select_dtypes([np.number]).columns
matriz = df_vif[numericos].corr()
```

```python
# grafico da nova matriz de correlacao
fig, ax = plt.subplots(figsize=(10,10))
ax.matshow(matriz)
plt.show()
```

```python
# lista de variaveis com maior correlacao com a variavel alvo
corr_pd = pd.DataFrame(matriz, columns=numericos)
corr_pd.index = numericos
np.abs(corr_pd['label']).sort_values(ascending=False)[:20]
```
