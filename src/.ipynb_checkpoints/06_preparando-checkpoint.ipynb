```python
import pandas as pd
import numpy as np
import pickle
from os import listdir
from os.path import isfile, join
```

```python
caminho = '../data/novo/csv/'
```

Recuperando lista de ids completos

```python
with open(caminho + 'ids.dat', 'rb') as fp:   # Unpickling
    ids = pickle.load(fp)
```

```python
arquivos = [f for f in listdir(caminho) if isfile(join(caminho, f)) and '.csv' in f]
```

Recuperando curriculos completos

```python
for arquivo in arquivos:
    nome = arquivo.replace('.csv', '')
    exec(nome + ' = pd.read_csv(caminho + "' + arquivo + '", low_memory=False)')
    exec(nome + ' = ' + nome + '.loc['+ nome +'.id.isin(ids)]')
```

```python
df = contactinfo[['id', 'contactinfo_contactmethod_postaladdress_region']]
df.is_copy = False
df.columns = ['id', 'localizacao']
df['localizacao'].fillna('br-df', inplace=True)
```

```python
temp = culture[['id', 'culture_country', 'culture_language']]
temp.columns = ['id', 'pais_origem', 'idioma_nativo']
df = df.merge(temp, how='left', on='id')
df.is_copy = False
print np.unique(df.isnull().any(axis=1))
df
```

```python
ls = []

for chave in ids:
    d = {'id': chave}
    idiomas = language.loc[language['id'] == chave]
    
    for index, idioma in idiomas.iterrows():
        i = idioma['language_languagecode']
        if idioma['language_read'] == True:
            exec('d["' + i + '_le"] = 1')
        if idioma['language_speak'] == True:
            exec('d["' + i + '_fala"] = 1')
        if idioma['language_write'] == True:
            exec('d["' + i + '_escreve"] = 1')
    ls.append(d)

temp = pd.DataFrame(ls)
temp.fillna(0, inplace=True)
temp
```

```python
df = df.merge(temp, how='left', on='id')
df.is_copy = False
print np.unique(df.isnull().any(axis=1))
df
```
